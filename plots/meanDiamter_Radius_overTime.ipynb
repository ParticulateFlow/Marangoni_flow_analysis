```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


folder = '../../../Data/results/dropletDiameter_radius_over_time/gap10/'
files = os.listdir(folder)
frameNumbers = [int(f.split('_')[0]) for f in files ]

dfs = []
for f in files:
    df = pd.read_csv(os.path.join(folder, f))
    dfs.append(df)
```

```python
sns.set()
fig, ax = plt.subplots()
for df, ind in zip(dfs, frameNumbers):
    df.plot(x='r',y='d', ax=ax, label=ind, xlabel='Radius', ylabel='Mean Diameter')
```
